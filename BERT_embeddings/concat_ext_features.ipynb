```python
import pickle as pkl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fnc_train_mean = pkl.load(open('/sbert_embeddings/fnc_train_mean.pkl','rb'))
fnc_train_mean2 = pkl.load(open('/sbert_embeddings/fnc_train_mean2.pkl','rb'))
```

```python
fnc_train_mean.reset_index(inplace=True)
```

```python
fnc_train_mean2.dropna(inplace=True)
fnc_train_mean2.reset_index(inplace=True)
```

```python
fnc_train_mean_ext = pd.concat([fnc_train_mean, fnc_train_mean2.iloc[:, -22:]], axis=1)
```

```python
pkl.dump(fnc_train_mean_ext, open('/sbert_embeddings/fnc_train_mean_ext.pkl','wb'))
```
